### 🧠 Model Building

In this section, we document the steps followed to build models for predicting employee performance ratings.

#### 1. Model Selection Rationale

Given that the target variable PerformanceRating is categorical, multiple classification models were considered to compare performance and robustness. These models included:

1. Logistic Regression: A simple, interpretable baseline model that performs well on linearly separable data.

2. Decision Tree Classifier: Captures non-linear relationships and provides visual interpretability.

3. Random Forest Classifier: An ensemble method that reduces overfitting and generally provides better accuracy than individual trees.

4. Support Vector Machine (SVM): Effective in high-dimensional spaces and useful for classification tasks with clear class margins.

5. K-Nearest Neighbors (KNN): A non-parametric method that classifies based on proximity to training samples.

6. Gradient Boosting (e.g., XGBoost): An advanced boosting algorithm that builds models sequentially to correct errors, often achieving high predictive performance.

These models were selected to balance:

Accuracy and generalization performance,

Computational cost, and

Interpretability, which is critical for HR decision-making.

#### 2. Assumptions

- The dataset is representative of the current employee population.
- The performance ratings are accurate and unbiased.
- Features such as promotion history, department, and experience impact performance outcomes.
- There is no data leakage from future information.

#### 3. Preprocessing Summary

Before modeling, the data underwent several preprocessing steps:

- **Categorical Encoding**: Variables like Department and Gender were encoded using label encoding or one-hot encoding as needed.
- **Feature Scaling**: Numerical features (e.g., Age, YearsAtCompany) were standardized for models sensitive to scale.
- **Train-Test Split**: Data was split into training (80%) and test (20%) sets to evaluate model generalization.

#### 4. Model Training

Each model was trained on the training data. Cross-validation was used for performance estimation and hyperparameter tuning where applicable.

#### 5. Evaluation Metrics

The models were evaluated using the following metrics:
- **Accuracy**: Overall correctness of predictions.
- **Precision, Recall, F1 Score**: Especially useful in understanding class-specific performance.
- **Confusion Matrix**: To visualize correct and incorrect predictions.
- **ROC-AUC** (if applicable): For assessing classification performance across thresholds.

Final model selection was based on performance and interpretability, aligning with the business need for actionable and transparent insights.
